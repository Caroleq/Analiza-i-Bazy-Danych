```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import plotnine as p9
```

```python
df = pd.read_csv('exercise.csv', nrows=40)
```

```python
results = smf.ols('y ~ x1 + x2 + x1 * x2', data=df).fit()
```

```python
wyn=results.params
```

```python
results.summary()
```

```python
# Display predictions.
# First line - slope is x1
# Second line - slope is x1 combined with wyn['x1:x2']
fig4=(p9.ggplot(p9.aes(x='x1',y='y',color='x2'),data=df)
      +p9.geom_point()
      +p9.geom_abline(p9.aes(intercept=wyn['Intercept'],slope=wyn['x1']))
      +p9.geom_abline(p9.aes(intercept=wyn['Intercept']+wyn['x1'],slope=wyn['x2']+wyn['x1:x2']))     )
```

```python
fig4
```

```python
# compute difference between predcted and real value
df['y_pred']=results.predict()
df['residuals']=df['y']-df['y_pred']
fig4_res=(p9.ggplot(p9.aes(x='x1',y='residuals',color='x2'),data=df)
      +p9.geom_point())
```
